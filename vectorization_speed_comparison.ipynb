```julia
using Gadfly

function unvect()
    x = rand(500)
    y = rand(500)
    a = rand()
    b = rand()
    d = Array(Float64, 0)
    twins = Array(Float64, 0,2)
    
    for i in 1:500
        d = [d; sqrt((x[i] - a)^2 + (y[i] - b)^2)]
        if d[end] < 0.05
            twins = [twins; [x y][end, :]]
        end
    end
    
    return twins
end

times = Array(Float64, 0)
num_trials = 300

for j = [1:num_trials]
    t0 = time();
    unvect()
    time_elapsed = (time()-t0) * 1000;
    push!(times, time_elapsed)
end

println("Mean time: $(mean(times)) msec")
set_default_plot_size(25cm, 15cm)
plot(x=[1:num_trials], y=times, Geom.point(), Geom.line(),
    Guide.xlabel("trial"), Guide.ylabel("time, msec"), Guide.title("Unvectorized"))

```

```julia
using Gadfly

function vect()
    x = rand(500)
    y = rand(500)
    a = rand()
    b = rand()
    
    d = sqrt((x-a).^2 + (y-b).^2)
    twins = [x y][d .< 0.05,:]
    
    return twins
end

times = Array(Float64, 0)
num_trials = 300

for j = [1:num_trials]
    t0 = time();
    vect()
    time_elapsed = (time()-t0) * 1000;
    push!(times, time_elapsed)
end

println("Mean time: $(mean(times)) msec")
set_default_plot_size(25cm, 15cm)
plot(x=[1:num_trials], y=times, Geom.point(), Geom.line(),
    Guide.xlabel("trial"), Guide.ylabel("time (ms)"), Guide.title("Using vectorization"))
```

```julia
using DataFrames, Gadfly

function df_distance()
    x = rand(500)
    y = rand(500)
    a = rand()
    b = rand()
    
    df = DataFrame(x=x, y=y)
       
    d = sqrt((df[:x]-a).^2 + (df[:y]-b).^2)
    twins = df[d .< 0.05,:]
    
    return twins
end

times = Array(Float64, 0)
num_trials = 300

for j = [1:num_trials]
    t0 = time();
    df_distance()
    time_elapsed = (time()-t0) * 1000;
    push!(times, time_elapsed)
end

println("Mean time: $(mean(times)) msec")
set_default_plot_size(25cm, 15cm)
plot(x=[1:num_trials], y=times, Geom.point(), Geom.line(),
    Guide.xlabel("trial"), Guide.ylabel("time (ms)"), Guide.title("Using DataFrames"))
```
