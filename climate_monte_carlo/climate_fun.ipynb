```julia
using Gadfly
```

```julia
function pop_temps(num_years, highest_temp)
    return 1:num_years, rand(num_years) * highest_temp
end

# test
years, temps = pop_temps(150, 110)
plot(x=years, y=temps, Geom.point(), Geom.line(), Scale.y_continuous(minvalue=0, maxvalue=120),
    Guide.ylabel("temperature"), # label for y-axis
    Guide.xlabel("year"),  # label for x-axis
    Guide.title("simulated temperatures"))  # a title
```

```julia
sort_ix = sortperm(temps, rev=true) # create indexes of temperatures, sorted in reverse order (high → low)
println("highest temps:")
println("year \t temp")
println([ sort_ix[1:10] temps[sort_ix][1:10] ]) # print years and temps corresponding to the top ten indexes
```

```julia
years, temps = pop_temps(150, 110)
sort_ix = sortperm(temps, rev=true)

# plotting stuff

set_default_plot_size(25cm, 15cm)

all_plot_layer = layer(x=years, y=temps, Geom.line(), Geom.point(), Theme(default_color=color("seagreen")))
max_plot = plot(x=sort_ix[1:10], y=temps[sort_ix][1:10], Geom.point(), Theme(default_color=color("indianred")),
            Scale.y_continuous(minvalue=0, maxvalue=120),
            Guide.ylabel("temperature"), # label for y-axis
            Guide.xlabel("year"),  # label for x-axis
            Guide.title("simulated temperatures: red = max 10"))  # a title)
recent_line = layer(x=[125, 125], y=[0, 120], Geom.line(), Theme(default_color=color("black")) )

append!(max_plot.layers, all_plot_layer)
append!(max_plot.layers, recent_line)

display(max_plot)
```
