```julia
using Gadfly
```

```julia
temp = rand(150) * 110
year_x = 1:150

plot(x=year_x, y=temp, Geom.point(), Geom.line())
```

```julia
sort_ix = sortperm(temp, rev=true)
println("highest temps:")
println("year \t temp")
println([ sort_ix[1:25] temp[sort_ix][1:25] ])
```

```julia
p = plot(x=year_x, y=temp, Geom.point(), Geom.line())
q = layer(x=sort_ix[1:25], y=temp[sort_ix][1:25])
append!(p, q)
```
